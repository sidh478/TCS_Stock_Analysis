```python
from pandas.plotting import lag_plot
for i in tcs.columns:
    lag_plot(tcs[i])
    plt.show()
```

From lag_plot,we can visualize the autocorrelation structure of the time series data.
1. In Close feature it suggests a strong positive autocorrelation
2. In Turnover it suggests no or weak autocorrelation.
3. In %Deliverable it suggests  Moderate co-realtion.


### Time series decomposition plot

```python
from statsmodels.tsa.seasonal import seasonal_decompose
for i in tcs.columns:
    decompose_ts_add = seasonal_decompose(tcs[i],period=12)
    decompose_ts_add.plot()
    plt.show()
```

### ACF and PACF plot

ACF - plot between lags and values of correlation

PACF - partial auto-correlation function, plot between lags and residuals 

(Basically instead of finding correlations of present with lags like ACF, it finds correlation of the residuals)


```python
import statsmodels.graphics.tsaplots as tsa_plots
for i in tcs.columns:
    tsa_plots.plot_acf(tcs[i],lags=12)
    tsa_plots.plot_pacf(tcs[i],lags=12)
    plt.show()
```
